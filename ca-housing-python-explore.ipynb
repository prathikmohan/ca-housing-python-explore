#### using groupby method

```python
grp_age=df.groupby(['housing_median_age']) #group data using housing median age from df and assign to grp_age
```

```python
grp_age.mean() #show mean of each column per each group in grp_age
```

```python
df.groupby(['housing_median_age'])['latitude'].mean() #calculate mean latitude for each housing median age group
```

```python
df.groupby(['housing_median_age'])[['latitude']].mean() #SAME as ABOVE with 2 brackets/output is dataframe - calculate mean latitude for each housing median age group
```

#### filtering

```python
df['latitude']<35 #filter latitude in boolean
```

```python
lat_less35 = df[df['latitude']<35] #output to data frame lat_less35
```

```python
lat_less35.groupby(['housing_median_age']).mean() #grouped by housing median age for < 35 latitude values
```
